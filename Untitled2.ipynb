```python
import os
import PIL
import collections
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
IMG_HEIGHT = 224
IMG_WIDTH = 224
N_CLASSES = 21

LIST_DIR = './data/VOCdevkit/VOC2012/ImageSets/Segmentation'
SEGMENTATION_DIR = "./data/VOCdevkit/VOC2012/SegmentationClass"
OUTPUT_DIR = "./data/VOCdevkit/VOC2012/SegmentationClassRaw"
```

```python
def _save_annotation(annotation, filename):
    """Saves the annotation as png file.
    Args:
    annotation: Segmentation annotation.
    filename: Output filename.
    """
    
    pil_image = PIL.Image.fromarray(annotation.astype(dtype=np.uint32))
    with tf.io.gfile.GFile(filename, mode='w') as f:
        pil_image.save(f, 'PNG')
    
#     with tf.io.gfile.GFile(filename, mode='w') as f:
#         pil_image.save(f, 'PNG')
        
def _remove_colormap(filename):
    """Removes the color map from the annotation.
    Args:
    filename: Ground truth annotation filename.
    Returns:
    Annotation without color map.
    """
    image = tf.io.gfile.GFile(filename, 'rb').read()
    image = tf.image.decode_png(image, channels=3)
    return image
    
def remove_color_map_process():
    # Create the output directory if not exists.
    if not tf.io.gfile.isdir(OUTPUT_DIR):
        tf.io.gfile.mkdir(OUTPUT_DIR)

    annotations = tf.io.gfile.glob(os.path.join(SEGMENTATION_DIR, '*.png'))
    for annotation in tqdm(annotations):
        raw_annotation = _remove_colormap(annotation)
        filename = os.path.basename(annotation)[:-4]
        plain_annotation = voc_label_indices(raw_annotation.numpy(), build_colormap2label())
        _save_annotation(plain_annotation, os.path.join(OUTPUT_DIR, filename + '.png'))

remove_color_map_process()
```

```python
def get_train_val():
    dataset_splits = tf.io.gfile.glob(os.path.join(LIST_DIR, "*.txt"))
    splits = collections.defaultdict(list)

    for dataset_split in dataset_splits:
        basename = os.path.basename(dataset_split)[:-4]
        filenames = [x.strip('\n') for x in open(dataset_split, 'r')]
        splits[basename] = filenames
        
    return splits
        
files = get_train_val()
```

```python
VOC_COLORMAP = [[0, 0, 0], [128, 0, 0], [0, 128, 0], [128, 128, 0],
                [0, 0, 128], [128, 0, 128], [0, 128, 128], [128, 128, 128],
                [64, 0, 0], [192, 0, 0], [64, 128, 0], [192, 128, 0],
                [64, 0, 128], [192, 0, 128], [64, 128, 128], [192, 128, 128],
                [0, 64, 0], [128, 64, 0], [0, 192, 0], [128, 192, 0],
                [0, 64, 128]]

# Saved in the d2l package for later use
VOC_CLASSES = ['background', 'aeroplane', 'bicycle', 'bird', 'boat',
               'bottle', 'bus', 'car', 'cat', 'chair', 'cow',
               'diningtable', 'dog', 'horse', 'motorbike', 'person',
               'potted plant', 'sheep', 'sofa', 'train', 'tv/monitor']

# Saved in the d2l package for later use
def build_colormap2label():
    """Build an RGB color to label mapping for segmentation."""
    colormap2label = np.zeros(256 ** 3)
    for i, colormap in enumerate(VOC_COLORMAP):
        colormap2label[(colormap[0]*256 + colormap[1])*256 + colormap[2]] = i
    return colormap2label

# Saved in the d2l package for later use
def voc_label_indices(colormap, colormap2label):
    """Map an RGB color to a label."""
    colormap = colormap.astype(np.int32)
    idx = ((colormap[:, :, 0] * 256 + colormap[:, :, 1]) * 256
           + colormap[:, :, 2])
    return colormap2label[idx]
```

```python
voc_dir = "./data/VOCdevkit/VOC2012/"

def read_voc_images(voc_dir, is_train=True):
    """Read all VOC feature and label images."""
    txt_fname = os.path.join(voc_dir, 'ImageSets', 'Segmentation',
                             'train.txt' if is_train else 'val.txt')
    with open(txt_fname, 'r') as f:
        images = f.read().split()
    features, labels = [], []
    for i, fname in enumerate(images):
        label = tf.io.gfile.GFile(os.path.join(voc_dir, 'SegmentationClass', '%s.png' % fname), 'rb').read()
        label = tf.image.decode_png(label, channels=3)
        labels.append(label)
    return features, labels

train_features, train_labels = read_voc_images(voc_dir, True)
```
